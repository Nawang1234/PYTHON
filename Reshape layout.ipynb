```python
#using in keyword for finding word
#groupby keyword combines all similar data

no2_subset = no2.sort_index().groupby(["location"]).head(2)
no2_subset
```

```python
#converting the values of three stations into columns

no2_subset.pivot(columns = "location", values = "value")
```

```python
air_quality.pivot_table(values = "value", 
                        index = "location", 
                        columns = "parameter", 
                        aggfunc = "mean")


#air_quality: This is your original DataFrame containing air quality data.
#pivot_table: This method is used to create a new DataFrame where:
#values="value": The data in the "value" column will be aggregated.
#index="location": Each unique location will be represented as a row in the pivot table.
#columns="parameter": Each unique parameter (like NO2, PM10, etc.) will be represented as a column.
#aggfunc="mean": The aggregation function used to summarize the data will be the mean. This means that for each location, the average value of each parameter will be calculated."""
```

```python
#combine two tables with different data...

import pandas as pd

# Load air quality NO2 data
air_quality_no2 = pd.read_csv("air_quality_no2.csv", parse_dates=True)

# Select specific columns
air_quality_no2 = air_quality_no2[["date.utc", "location", "parameter", "value"]]

# Load air quality PM2.5 data
air_quality_pm25 = pd.read_csv("air_quality_pm25.csv", parse_dates=True)

# Select specific columns
air_quality_pm25 = air_quality_pm25[["date.utc", "location", "parameter", "value"]]

# Concatenate the two DataFrames
air_quality = pd.concat([air_quality_pm25, air_quality_no2], axis=0)

# Display the first few rows of the combined DataFrame
print(air_quality)
```
